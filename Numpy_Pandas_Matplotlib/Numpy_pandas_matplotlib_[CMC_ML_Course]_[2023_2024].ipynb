Обратите внимание, что один и тот же ролик может быть опубликован в нескольких странах, и ему может соответствовать несколько строк в таблице. На графике необходимо учесть каждый ролик только 1 раз.

```python
# Your code here
vids = data[['video_id', 'publishedAt']].groupby(['video_id', 'publishedAt']).nunique()
vids = vids.index.to_frame(index=False)['publishedAt'].dt.date.value_counts().sort_index()

plt.figure(figsize=(10,5))
plt.title('Зависимость количества выпущенных роликов от даты публикации')
plt.ylabel('Количество выпущенных роликов')
plt.xlabel('Дата публикации')
plt.plot(vids, marker='.', color='red', mec='black')
plt.grid(True)
plt.xticks(vids.index, rotation=45, ha='right')
plt.locator_params (axis='x', tight=True, nbins= 30)
```

```python
import time
import numpy as np

import functions_vectorised
import functions
```

```python
# генерация данных

sum_non_neg_diag_data = []
for n in [10, 100, 1000, 10000]:
    X = np.random.randint(0, 3, (n, n))
    sum_non_neg_diag_data += [X]

are_multisets_equal_data = []
for n in [10, 100, 1000, 10000]:
    x = np.random.randint(0, 2, n)
    y = np.random.randint(0, 2, n)
    are_multisets_equal_data += [(x, y)]

max_prod_mod_3_data = []
for n in [10, 100, 1000, 10000]:
    x = np.random.randint(1, 100, n)
    max_prod_mod_3_data += [x]

convert_image_data = []
for n in [10, 100, 1000]:
    image = np.random.random((n, n, 10))
    weights = np.random.random(10)
    convert_image_data += [(image, weights)]

rle_scalar_data = []
for n in [10, 100, 1000, 10000]:
    x = np.random.randint(1, 3, (n, 2))
    y = np.random.randint(1, 3, (n, 2))
    shapes = np.random.randint(1, 3, n)
    x[:, 1] = shapes
    y[:, 1] = shapes

    rle_scalar_data += [(x, y)]

cosine_distance_data = []
for n in [10, 100]:
    X = np.random.random((n, n))
    Y = np.random.random((n, n))
    cosine_distance_data += [(X, Y)]
```

```python
# функция для подсчета времени работы

def count_time(func, *data):
    tick = time.perf_counter()
    func(*data)
    return time.perf_counter() - tick
```

```python
# получим время работы функций

# sum_non_neg_diag

sum_non_neg_diag_time = []
sum_non_neg_diag_vectorised_time = []

for X in sum_non_neg_diag_data:
  sum_non_neg_diag_time += [count_time(functions.sum_non_neg_diag, X.tolist())]
  sum_non_neg_diag_vectorised_time += [count_time(functions_vectorised.sum_non_neg_diag, X)]

# are_multisets_equal

are_multisets_equal_time = []
are_multisets_equal_vectorised_time = []

for x, y in are_multisets_equal_data:
  are_multisets_equal_time += [count_time(functions.are_multisets_equal, x.tolist(), y.tolist())]
  are_multisets_equal_vectorised_time += [count_time(functions_vectorised.are_multisets_equal, x, y)]

# max_prod_mod_3

max_prod_mod_3_time = []
max_prod_mod_3_vectorised_time = []

for x in max_prod_mod_3_data:
  max_prod_mod_3_time += [count_time(functions.max_prod_mod_3, x.tolist())]
  max_prod_mod_3_vectorised_time += [count_time(functions_vectorised.max_prod_mod_3, x)]

# convert_image

convert_image_time = []
convert_image_vectorised_time = []

for image, weights in convert_image_data:
  convert_image_time += [count_time(functions.convert_image, image.tolist(), weights.tolist())]
  convert_image_vectorised_time += [count_time(functions_vectorised.convert_image, image, weights)]

# rle_scalar

rle_scalar_time = []
rle_scalar_vectorised_time = []

for x, y in rle_scalar_data:
  rle_scalar_time += [count_time(functions.rle_scalar, x.tolist(), y.tolist())]
  rle_scalar_vectorised_time += [count_time(functions_vectorised.rle_scalar, x, y)]

# cosine_distance

cosine_distance_time = []
cosine_distance_vectorised_time = []

for X, Y in cosine_distance_data:
  cosine_distance_time += [count_time(functions.cosine_distance, X.tolist(), Y.tolist())]
  cosine_distance_vectorised_time += [count_time(functions_vectorised.cosine_distance, X, Y)]
```

```python
# ВАШИ ГРАФИКИ ЗДЕСЬ
n = [10, 100, 1000, 10000]
plt.figure(figsize=(20, 10))

# sum_non_neg_diag

plt.subplot(2, 3, 1)
plt.title('Сравнение времени работы реализаций от размера данных: sum_non_neg_diag', fontsize = 8)
plt.ylabel('Время выполнения')
plt.xlabel('Размер данных')
plt.plot(n, sum_non_neg_diag_time, c='r')
plt.plot(n, sum_non_neg_diag_vectorised_time, c='g')
plt.legend(('sum_non_neg_diag', 'sum_non_neg_diag_vectorised'))
plt.grid(True)

# are_multisets_equal

plt.subplot(2, 3, 2)
plt.title('Сравнение времени работы реализаций от размера данных: are_multisets_equal', fontsize = 8)
plt.ylabel('Время выполнения')
plt.xlabel('Размер данных')
plt.plot(n, are_multisets_equal_time, c='r')
plt.plot(n, are_multisets_equal_vectorised_time, c='g')
plt.legend(('are_multisets_equal', 'are_multisets_equal_vectorised'))
plt.grid(True)

# max_prod_mod_3

plt.subplot(2, 3, 3)
plt.title('Сравнение времени работы реализаций от размера данных: max_prod_mod_3', fontsize = 8)
plt.ylabel('Время выполнения')
plt.xlabel('Размер данных')
plt.plot(n, max_prod_mod_3_time, c='r')
plt.plot(n, max_prod_mod_3_vectorised_time, c='g')
plt.legend(('max_prod_mod_3', 'max_prod_mod_3_vectorised'))
plt.grid(True)

# convert_image

plt.subplot(2, 3, 4)
plt.title('Сравнение времени работы реализаций от размера данных: convert_image', fontsize = 8)
plt.ylabel('Время выполнения')
plt.xlabel('Размер данных')
plt.plot(n[:3], convert_image_time, c='r')
plt.plot(n[:3], convert_image_vectorised_time, c='g')
plt.legend(('convert_image', 'convert_image_vectorised'))
plt.grid(True)

# rle_scalar

plt.subplot(2, 3, 5)
plt.title('Сравнение времени работы реализаций от размера данных: rle_scalar', fontsize = 8)
plt.ylabel('Время выполнения')
plt.xlabel('Размер данных')
plt.plot(n, rle_scalar_time, c='r')
plt.plot(n, rle_scalar_vectorised_time, c='g')
plt.legend(('rle_scalar', 'rle_scalar_vectorised'))
plt.grid(True)

# cosine_distance

plt.subplot(2, 3, 6)
plt.title('Сравнение времени работы реализаций от размера данных: cosine_distance', fontsize = 8)
plt.ylabel('Время выполнения')
plt.xlabel('Размер данных')
plt.plot(n[:2], cosine_distance_time, c='r')
plt.plot(n[:2], cosine_distance_vectorised_time, c='g')
plt.legend(('cosine_distance', 'cosine_distance_vectorised'))
plt.grid(True)
```
